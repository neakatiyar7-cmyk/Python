```python
# when condition is true

n1 = int(input("Enter first number"))
n2 = int(input("Enter Second number"))
if(n1>n2):
          print(f"{n1} is greater than {n2}")
          print("if block")
print("Hello")
```

```python
# when condition is  not true

n1 = int(input("Enter first number"))
n2 = int(input("Enter Second number"))
if(n1>n2):
          print(f"{n1} is greater than {n2}")
          print("if block")
print("Hello")
```

```python
# When multiple conditions to check and print the output

n1 = int(input("Enter first number"))
n2 = int(input("Enter Second number"))
if(n1>n2):
          print(f"{n1} is greater than {n2}")
          print("if block")
elif(n1>n2):
          print(f"{n2} is greater than {n1}")
          print("elif block")
else:
          print(f"{n2} is equal {n1}")
          print("else block")
print("Hello")
```

```python
#Take 5 subject marks from user  and display total marks
#percentage and based on percentage display grade(A+,A,B,C)

n1 = int(input("Enter any number"))
n2 = int(input("Enter any number"))
n3 = int(input("Enter any number"))
n4 = int(input("Enter any number"))
n5 = int(input("Enter any number"))

totalmarks = n1+n2+n3+n4+n5
per = (totalmarks/500)*100
print(f" Total marks of a student is {totalmarks}")
print(f" Percentage of a student is {per}")

if(per>=90):
    print(f" percentage is {per} so grade is A+")
elif(per>=80 and per<90):
    print(f" percentage is {per} so grade is A")
elif(per>=70 and per<80):
       print(f" percentage is {per} so grade is B")
elif(per>=60 and per<70):
       print(f" percentage is {per} so grade is C")
else:
    print(f" percentage is below {per} so grade is D")
```

```python
# print 10 to 1

i =10

while(i>=1):
    print(i)
    i=i-1
print("complete")
print(f"last value of i is {i}")
```

```python
# print multiplication table of user given number 
i=1 

num = int(input("Enter the number"))

while(i<=10):
      table1 = num * i
      print(f" the table of {num} is {num}*{i} = {table1}")
      i=i+1
print("table is completed")
```
