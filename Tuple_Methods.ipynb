```python
# tuples are immutable sequences, meaning they cannot be changed after creation.
#Because of this immutability, the tuple class provides only two built-in methods.

l3 =(10,20,50,30,40,50,20)
```

```python
#count(value): Returns the number of times a specified value occurs in the tuple.

l3.count(20)
```

```python
#ndex(value, [start, [end]]): Searches the tuple for a specified value and returns the index of its first occurrence.
 
l3.index(50)
```

```python
# len(tuple): Returns the total number of items in the tuple.

len (l3)
```

```python
# max(tuple) / min(tuple): Returns the largest or smallest item, provided all elements are of comparable types.


max(l3)
```

```python
min(l3)
```

```python
#sum(tuple): Returns the arithmetic sum of all numeric elements.

sum(l3)
```

```python
#sorted(tuple): Returns a new sorted list containing all items from the tuple (the original tuple remains unchanged).

sorted(l3)
```

```python
# any(tuple) / all(tuple): any() returns True if at least one element is true; all() returns True only if every element is true

any(l3)
```

```python
all(l3)
```
