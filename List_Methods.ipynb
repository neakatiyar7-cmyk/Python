```python
L3=[10,20,30,40,50]
```

```python
#append :adds the element at the last of the list and can only take one argument

L3.append(100)
L3
```

```python
# insert : inserts the vale at any index as per user requirement

L3.insert(3,2000)
L3
```

```python
#extend : 	Adds multiple items from an iterable like list, tuple, string, set, or dictionary keys

L3.extend([20000,30000,4000])   # adding list 

Lists: [1, 2].extend([3, 4]) results in [1, 2, 3, 4].
Tuples: [1, 2].extend((3, 4)) results in [1, 2, 3, 4].
Strings: ['a'].extend('bc') results in ['a', 'b', 'c'] (it treats the string as a sequence of characters).
```

```python
# pop :the pop() method is used to remove an element from a list and return its value.
# This method modifies the original list in-place

L3.pop()
L3
```

```python
# remove :In Python, you can remove items from a list using several built-in methods 
            # depending on whether you want to delete by value, index, or condition. 

L3.remove(100)
L3
```

```python
#del :Deletes the item at a specific index but does not return it.

del L3[5]
```

```python
# clear(): Empties the entire list, leaving it as []
L3.clear()
```
